```python
import pandas as pd
```

```python
train_df = pd.read_csv("../content/gdrive/MyDrive/ML/perceptron-train.csv", header=None)
```

```python
train_df.shape
```

```python
train_df.head(5)
```

```python
test_df = pd.read_csv("../content/gdrive/MyDrive/ML/perceptron-test.csv", header=None)
```

```python
test_df.shape
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
X_train = train_df.iloc[:,1:]
y_train = train_df.iloc[:,0:1]
X_test = test_df.iloc[:,1:]
y_test = test_df.iloc[:,0:1]
```

```python
model_1 = Perceptron(random_state=241)
model_1.fit(X_train, y_train)
```

```python
model_1.predict([[-0.866495, 1191.232457], [0.866495, 500.232457]])
```

```python
accuracy_1 = accuracy_score(y_pred=model_1.predict(X_test), y_true=y_test)
accuracy_1
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_2 = Perceptron(random_state=241)
model_2.fit(X_train_scaled, y_train)
```

```python
accuracy_2 = accuracy_score(y_pred=model_2.predict(X_test_scaled), y_true=y_test)
accuracy_2
```

```python
accuracy_2 - accuracy_1
```
