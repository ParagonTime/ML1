```python
data = sklearn.datasets.load_boston()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import scale
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
data.data.shape
```

```python
data.target.shape
```

```python
result = pd.DataFrame(np.zeros((20,10)))

for i in range(20):
    i+=1

    for j in range(10):
        j+=1

        KNR = KNeighborsRegressor(n_neighbors=i,
                                weights='distance',
                                p=j, metric='minkowski')

        kf = KFold(n_splits=5, random_state=42, shuffle=True)

        for dataX in kf.split(data.data):
            KNR.fit(scale(data.data[dataX[0]]), data.target[dataX[0]])

        vall = cross_val_score(KNR, scale(data.data[dataX[1]]), data.target[dataX[1]],  scoring='neg_mean_squared_error')
        #print('neighbors =', i, ' p =', j, 'vall =', vall.mean())
        result.iloc[i-1,j-1] = (vall.mean())
```
