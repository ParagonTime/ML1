```python
data = sklearn.datasets.load_boston()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import scale
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
data.data.shape
```

```python
data.target.shape
```

```python
KNR = KNeighborsRegressor(n_neighbors=5,
                          weights='distance',
                          p=9, metric='minkowski')

kf = KFold(n_splits=2, random_state=42, shuffle=True)

for dataX in kf.split(data.data):

    KNR.fit(scale(data.data[dataX[0]]), data.target[dataX[0]])

    vall = cross_val_score(KNR, scale(data.data[dataX[1]]), data.target[dataX[1]],  scoring='neg_mean_squared_error')
    print(vall.mean())
```
