```python
import pandas as pd
```

```python
import re
```

```python
data = pd.read_fwf("../content/gdrive/MyDrive/ML/wine.data")
```

```python
pattern = r"[\w]*\.*\w*"
buf = [[]]
```

```python
for i in data.values:
    allinc = re.findall(pattern, i[0])[::2]
    buf.append(allinc)
```

```python
data2 = pd.DataFrame(buf)
```

```python
data2 = data2.dropna()
```

```python
data2
```

```python
data2.loc[:,1:13]
```

```python
data2.loc[:,0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
```

```python
neigh = KNeighborsClassifier(n_neighbors=1)

kf = KFold(n_splits=5, random_state=42, shuffle=True)

for dataX, dataY in kf.split(data2):
    
    neigh.fit(data2.loc[dataX+1, 1:], data2.loc[dataX+1, 0])

    vall = cross_val_score(neigh, data2.loc[dataY+1, 1:], data2.loc[dataY+1, 0])
    print(vall.mean())
```

```python
data3 = scale(data2)

neigh_scale = KNeighborsClassifier(n_neighbors=3)

kf = KFold(n_splits=5, random_state=42, shuffle=True)

for dataX, dataY in kf.split(data3):

    neigh_scale.fit(data3[dataX, 1:], data2.loc[dataX+1, 0])
    
    vall = cross_val_score(neigh_scale, data3[dataY, 1:], data2.loc[dataY+1, 0])
    print(vall.mean())
```

```python
pred = neigh.predict(data2.loc[:,1:])
```

```python
print(pred)
```

```python
pred = neigh_scale.predict(scale(data2.loc[:,1:]))
print(pred)
```
