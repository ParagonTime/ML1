```python
accurasy = accuracy_score(y_true=data_clas['true'], y_pred=data_clas['pred'])
precision = precision_score(y_true=data_clas['true'], y_pred=data_clas['pred'])
recall = recall_score(y_true=data_clas['true'], y_pred=data_clas['pred'])
f1_metric = f1_score(y_true=data_clas['true'], y_pred=data_clas['pred'])
print('accurasy = ', accurasy)
print('precision = ', precision)
print('recall = ', recall)
print('f1_metric = ', f1_metric)
```

```python
data_scores = pd.read_csv('/content/gdrive/MyDrive/ML/scores.csv')
```

```python
roc_logreg = roc_auc_score(y_true=data_scores['true'], y_score=data_scores['score_logreg'])
roc_svm = roc_auc_score(y_true=data_scores['true'], y_score=data_scores['score_svm'])
roc_knn = roc_auc_score(y_true=data_scores['true'], y_score=data_scores['score_knn'])
roc_tree = roc_auc_score(y_true=data_scores['true'], y_score=data_scores['score_tree'])
print('roc_logreg', roc_logreg)
print('roc_svm', roc_svm)
print('roc_knn', roc_knn)
print('roc_tree', roc_tree)
```
