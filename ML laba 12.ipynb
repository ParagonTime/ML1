```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/ML/abalone.csv')
```

```python
data.head(5)
```

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (0 if x == 'I' else -1))
```

```python
data.head(5)
```

```python
dataSigns = data.loc[:,'Sex':'ShellWeight']
print(dataSigns.head(5))
```

при 21 дереве модель показывает качество выше 0.52 \
при увеличении числа деревьев качество повышается

```python
n_estimators_list = []
cross_val_score_list = []

kf = KFold(n_splits=5, shuffle=True, random_state=1)


for i in range(1,51,1):
    model = RandomForestRegressor(n_estimators=i, random_state=1)
    val = cross_val_score(model, dataSigns, data['Rings'], cv=kf, scoring=make_scorer(r2_score))
   
    n_estimators_list.append(i)
    cross_val_score_list.append(val.mean())
```

```python
for n_estimators, val in zip(n_estimators_list, cross_val_score_list):
    print(n_estimators, f'{val:.3}')
```
