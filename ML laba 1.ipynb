```python
import re
```

```python
pattern = r'Miss\..(\w+)\b'
string = 'Johnston, Miss. Catherine Helen "Carrie"'

result = re.findall(pattern, string)
print(result) # работает
```

```python
pattern = r'Mrs\..+\((\w+)\b'
string = 'Shelley, Mrs. William (Imanita Parrish Hall)'

result = re.findall(pattern, string)
print(result) # работает
```

```python
pattern = r'Mrs\..+\((\w+)\b|Miss\..(\w+)\b'
string = 'Graham, Miss. Margaret Edith'

result = re.findall(pattern, string)
print(result) # работает
```

```python
from pandas.io.formats.style_render import DataFrame
data3 = []
```

```python
for i in data2:
    result = re.findall(pattern, i) # pattern = r'Mrs\..+\((\w+)\b|Miss\..(\w+)\b'
    if result:
        if result[0][0] != '':
            data3.append(result[0][0])
        else:
            data3.append(result[0][1])
```

```python
data4 = pd.DataFrame(data3, columns=['Name'])
```

```python
import numpy as np
zer = np.zeros(len(data4.Name.unique()))
```
