```python
data = pd.read_csv("../content/gdrive/MyDrive/ML/titanic.csv", index_col="PassengerId")
```

```python
data3 = data[['Survived','Pclass', 'Sex', 'Age', 'Fare']].dropna()
```

```python
dataSurv = data3[['Survived']]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data3['Sex'] = data3['Sex'].apply(lambda x: x == 'male')
```

```python
data3
```

```python
data2 = data3[['Pclass', 'Sex', 'Age', 'Fare']]
```

```python
data2
```

```python
clf = DecisionTreeClassifier(random_state=241)
```

```python
clf.fit(data2, dataSurv)
```

```python
importances = clf.feature_importances_
```
